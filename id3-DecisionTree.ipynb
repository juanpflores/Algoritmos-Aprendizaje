```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df1 = pd.DataFrame.from_csv("./data/Ejemplo12.csv", header=0)
df1
```

```python
# Conseguimos los features y targets de cada columna del DataFrame
tags = []
targets = []
features = []
for i in range(len(df1.columns) - 1):
    features.append(df1[df1.columns[i]].unique())
    tags.append(df1[df1.columns[i]].unique())
    print("Categoria {0}: {1} Valores: {2}".format(i, df1.columns[i], features[i]))

target = df1[df1.columns[-1]].unique()
tags.append(df1[df1.columns[-1]].unique())
print("Lista de Targets: {0}".format(target))
```

```python
def tag2numeric(tag, location):
    '''Convierte los tags del DataFrame a valores num√©ricos dependiedo del tag
    
    Args
    ------
    tag -- string con el nombre de la categoria
    location -- columna a la que pertenece el tag
    '''
    
    for i in range(len(tags[location])):
        if tag == tags[location][i]:
            return i
```

```python
# Ejecutamos tag2numeric para pasar de variables categoricas a enteros.

for i in range(len(df1.columns)):
    df1[df1.columns[i]] = df1[df1.columns[i]].apply(lambda x: tag2numeric(x,i))

df1
```

```python
# Separamos ambos en features y targets asi como training y test data
X_train, X_test, y_train, y_test = train_test_split(df1[df1.columns[:-1]], df1[df1.columns[-1]], test_size=0.2)
```

```python
# Para entrenar con ID3 debemos especificar que el criterio a seguir es la entropia
clf = DecisionTreeClassifier(criterion="entropy")
clf = clf.fit(X_train,y_train)
```

```python
prediccion = clf.predict(df1[df1.columns[:-1]])
```

```python
df_predict = df1.assign(prediccion=prediccion)
```

```python
print("El score de nuestro algoritmo es: {0}%".format(clf.score(X_test,y_test)*100))
```
